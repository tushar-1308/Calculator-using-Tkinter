```python
from tkinter import *

def click(event):
    global scvalue
    text = event.widget.cget("text")
    if text == "=":
        if strval.get().isdigit():
            value = int(strval.get())
        else:
            try:
                value = eval(screen.get())

            except Exception as e:
                print(e)
                value = "Error"


        strval.set(value)
        screen.update()

    elif text == "C":
        strval.set("")
        screen.update()

    else:
        strval.set(strval.get() + text)
        screen.update()

root = Tk()
root.geometry("480x650")
root.title("Calculator By Tushar Bansal")
# root.wm_iconbitmap("1.ico")

strval = StringVar()
strval.set("")
screen = Entry(root, textvar=strval, font="lucida 25 bold")
screen.pack(fill=X, ipadx=8, pady=10, padx=10)

f1 = Frame(root)
b = Button(f1, text="9", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f1, text="8", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f1, text="7", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f1, text="*", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)
f1.pack()


f2 = Frame(root)
b = Button(f2, text="6", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f2, text="5", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f2, text="4", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f2, text="-", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)
f2.pack()


f3 = Frame(root)
b = Button(f3, text="3", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f3, text="2", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f3, text="1", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f3, text="+", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)
f3.pack()


f4 = Frame(root)
b = Button(f4, text="0", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f4, text="00", padx=7, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=18, pady=12)
b.bind("<Button-1>", click)

b = Button(f4, text=".", padx=19, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=9, pady=12)
b.bind("<Button-1>", click)

b = Button(f4, text="/", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)
f4.pack()


f5 = Frame(root)
b = Button(f5, text="C", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f5, text="%", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

b = Button(f5, text="=", padx=15, pady=15, font="lucida 20 bold")
b.pack(side=LEFT, padx=12, pady=12)
b.bind("<Button-1>", click)

f5.pack()

root.mainloop()
```
